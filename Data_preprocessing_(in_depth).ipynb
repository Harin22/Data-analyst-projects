```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**importing datas:**



```python
df = pd.read_csv("amazon ds.csv")
display(df.head())
```

**Checked the missing file**

```python
display(pd.isna(df).sum())
```

**Checking the data types of the columns, before before proceeding with data cleaning and transformation.**

```python
display(df.dtypes)
```

**Convert non numerical to numerical columns**

```python
numeric_columns = ['rating', 'discount_percentage', 'actual_price', 'discounted_price']

for column in numeric_columns:
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace(r'[₹,%]', '', regex = True)
    df[column] = df[column].replace(',','')
    df[column] = pd.to_numeric(df[column], errors='coerce')

display(df.dtypes)
```

*everythings except "rating_count" is converted into Int or Float. thats because, "rating_count" has 2 null value.*

*there are several ways to handle null values (ex: drop, fill manually, mean, medium, mode, predict) instead of dropping, we can use medium*

```python

mean_rating_count = df['rating_count'].mean() #Filling with Mean
df['rating_count'] = df['rating_count'].fillna(mean_rating_count)
display(df.dtypes)

#median_rating_count = df['rating_count'].astype(float).median() #also Filled with Median
#df['rating_count'] = df['rating_count'].fillna(median_rating_count)
```

```python
#check duplicate values

duplicate_values = df.duplicated()
display(duplicate_values)

if duplicate_values.any():
  print("Yes there is duplicate values")
else:
  print("No there is no duplicate values")
```

```python
#finding if there is any outlier values

import seaborn as sns

values = ['rating', 'discount_percentage', 'actual_price', 'discounted_price', 'rating_count']
for v in values:
  sns.boxplot(df[v])
  plt.title("Boxplot of the values")
  plt.show()

```
