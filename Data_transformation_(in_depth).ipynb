```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numeric_cols = ['actual_price', 'discounted_price' ]

new_df[numeric_cols] = scaler.fit_transform(new_df[numeric_cols])
display(new_df[numeric_cols])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  # applying standardization
new_df[['rating', 'rating_count']] = scaler.fit_transform(new_df[['rating','rating_count']])

display(new_df[['rating', 'rating_count']])
```

**Feature engineering**

```python
new_df['final_price'] = new_df['actual_price'] - new_df['discounted_price']
display(new_df['final_price'])
```

checking if teh operation is done correctly on the scaled data.

```python
new_df['scaled_actual_price'] = new_df['actual_price']
new_df['scaled_discounted_price'] = new_df['discounted_price']

# Calculate final price on the scaled data
new_df['final_price'] = new_df['scaled_actual_price'] - new_df['scaled_discounted_price']
display(new_df['final_price'])
```
